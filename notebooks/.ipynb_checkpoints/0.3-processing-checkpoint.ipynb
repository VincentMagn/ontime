```python
import pandas as pd
import numpy as np
import ontime as on
```

## Creation of TimeSeries

```python
def generate_random_time_series(start, end, freq='D'):
    date_rng = pd.date_range(start=start, end=end, freq=freq)
    random_data = np.random.randn(len(date_rng))
    ts = pd.DataFrame(random_data, index=date_rng, columns=['values'])
    return ts
```

```python
ts = generate_random_time_series('2022-01-01', '2022-12-31')
```

```python
ts = on.TimeSeries.from_dataframe(ts)
```

```python
ts
```

## Apply fn along the normal timeline

with Lambda function

```python
add_two = on.processors.mapper(lambda x : x + 2)
new_ts = add_two.process(ts)
new_ts - ts
```

with normal function

```python
def add_2(x):
    return x + 2

add_two = on.processors.mapper(add_2)
new_ts = add_two.process(ts)
new_ts - ts
```

## Fn to split into windows

```python
mean = on.processors.windower({
    'function': 'mean',
    'mode': 'rolling',
    'window': 10
})
```

```python
new_ts = mean.process(ts)
```

```python
ts.plot()
```

```python
new_ts.plot()
```
